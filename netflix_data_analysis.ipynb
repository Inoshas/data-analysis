```python
#  Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

Data Analysis begin here :

```python
data=pd.read_csv('netflix_titles.csv')
size_of_data=print(data.shape[0])
data.columns
```

Plot data based on the type and the release year

```python
print(f"This contain data for {data['type'].unique()}")
#print(data['listed_in'])

category_movies = data[(data['listed_in'].str.contains('International')) & (data['type'] == 'Movie')]
category_TV_shows= data[(data['listed_in'].str.contains('International')) & (data['type'] == 'TV Show')]

print(f"Number of movies and TV Shows listed as International are {len(category_movies)} and {len(category_TV_shows)} respectively")


```

```python
#  Separate movie and TV series data
movies = data[data['type'] == 'Movie']
tv_shows = data[data['type'] == 'TV Show']


# Group the data by release year 
movies_by_year = movies['release_year'].values
tv_shows_by_year = tv_shows['release_year'].values
print(movies_by_year)

fig_comparison,ax_comparison=plt.subplots(2,1,figsize=(4,6))
plt.subplots_adjust(hspace=0.5) 
#Plot the histograms

ax_comparison[0].hist(movies_by_year,bins=100)
ax_comparison[1].hist(tv_shows_by_year,bins=100)

#plt.bar(movies_by_year, movies_by_year.values, alpha=0.7, label='Movies')
#plt.bar(tv_shows_by_year, tv_shows_by_year.values, alpha=0.7, label='TV Shows')
ax_comparison[0].set_xlabel('Release Year')
ax_comparison[1].set_xlabel('Release Year')

ax_comparison[0].set_ylabel('Movies')
ax_comparison[1].set_ylabel('TV shows')

ax_comparison[0].set_title('Number of Movies and TV Shows Released in  Each Year')

```
